```python
import pymysql
import pandas as pd

from dotenv import load_dotenv
from os import getenv
```

```python
load_dotenv(dotenv_path = '.env')
```

```python
getenv('host')
```

```python
conn = pymysql.connect(host = getenv('host'),
                       port = int(getenv('port')),
                       user = getenv('user'),
                       passwd = getenv('password'),
                       db = 'telcomarketing')
```

```python
baskets = pd.read_sql('select * from Baskets', conn)
baskets.to_csv('baskets.csv', index = False)
baskets
```

```python
customers = pd.read_sql('select * from Customers', conn)
customers.to_csv('customers.csv', index = False)
customers
```

```python
products = pd.read_sql('select * from Products', conn)
products.to_csv('products.csv', index = False)
products
```

```python
stores = pd.read_sql('select * from Stores', conn)
stores.to_csv('stores.csv', index = False)
stores
```

```python
timestamps = pd.read_sql('select * from Timestamps', conn)
timestamps.to_csv('timestamps.csv', index = False)
timestamps
```

```python
transactions = pd.read_sql('select * from Transactions', conn)
transactions.to_csv('transactions.csv', index = False)
transactions
```

```python
telco = pd.read_sql('select * from customer', conn)
telco.to_csv('telco.csv', index = False)
telco
```

```python
#Close connection
conn.close()
```
